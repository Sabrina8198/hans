# HeapSort
* 一開始的個人理解
  * 先將LIST依序排入二元樹中(詳見流程圖)
  * 由父節點和子節點相比，如果子節點較大，就要和父節點交換
  * 沒有子節點就不用換
  * 層層換完後就能完成堆積
  * 接著排序
  * 堆積最大的和最後一個元素交換，相比較完，最大的拿出來放在LIST最後
  * 再重複做一樣的動作，直到所有值都丟到排序後的LIST裡就完成啦
  
  * 參考資料
    * 概念理解 : https://magiclen.org/heap-sort/
    * GeeksforGeeks : https://www.geeksforgeeks.org/heap-sort/
    * youtube影片 : https://www.youtube.com/watch?v=GnKHVXv_rlQ&t=334s
    
* HeapSort
  * 若以升序排序說明，把陣列轉換成最大堆積(Max-Heap Heap)，這是一種滿足最大堆積性質(Max-Heap Property)的二元樹：對於除了根之外的每個節點i, A[parent(i)] ≥ A[i]。重複從最大堆積取出數值最大的結點(把根結點和最後一個結點交換，把交換後的最後一個結點移出堆)，並讓殘餘的堆積維持最大堆積性質。
  * 有分成MAX和MIN兩種

```python
def heapify(arr ,index, size):
    largest = index #設index為最大父節點
    rightindex = 2 * index + 2 #父節點index的右子節點在位置(2index+2)
    leftindex = 2 * index + 1 #父節點index的左子節點在位置(2index+1)
    if arr[leftindex] > arr[largest]:
        largest = leftindex
    if arr[rightindex] > arr[largest]:
        largest = rightindex
```

```python
def heap_sort(arr):
    n = len(arr) #設n為list的長度
    for i in range(n//2-1, -1, -1):
        heapify(arr, i, n)
    for i in range(n-1, 0, -1):
        arr[0], arr[i] = arr[i], arr[0]
        heapify(arr, 0, i)
    return arr
```

```python
def compare(a, b):
    if a < b:
        return -1
    if a > b:
        return 1 #這邊設定一個用來比較的參數compare，如果帶入的值a<b，回傳-1，a>b回傳1
```

```python
def heapify(arr ,index, size, compare):
    largest = index #設index為最大父節點
    rightindex = 2 * index + 2 #父節點index的右子節點在位置(2index+2)
    leftindex = 2 * index + 1 #父節點index的左子節點在位置(2index+1)
    if leftindex < size and compare(arr[leftindex], arr[largest]) > 0:
        largest = leftindex #如果左邊的子節點存在，運用compare function表較，>0代表a>b，左邊的子節點則變成父節點
    if rightindex < size and compare(arr[rightindex], arr[largest]) > 0:
        largest = rightindex #如果右邊的子節點存在，運用compare function表較，>0代表a>b，右邊的子節點則變成父節點
    if largest != index:
        arr[largest], arr[index] = arr[index], arr[largest] #如果原本設定的父節點不是最大的，就互換位置
        heapify(arr, largest, size, compare)
```

```python
def heap_sort(arr):
    n = len(arr) #設n為list的長度
    for i in range(n - 1, -1, -1): #從最後一個數開始n-1,n-2,n-3.........直到0
        heapify(arr, i, n, compare) #i值一直換，丟入heapify function裡
    for i in range(n - 1, 0, -1): #從最後面開始跑迴圈(尾，頭，往前推一個)
        arr[0], arr[i] = arr[i], arr[0] #這邊是要排序，拿出去值後，第一個數要和最後一個數互換，繼續排列，直到排完，所以也是跑迴圈
        heapify(arr, 0, i, compare)
    return arr
```

## 最後將程式碼轉為助教要求的格式

```python
class Solution(object):
    def heap_sort(self, nums):
        n = len(nums) #設n為list的長度
        for i in range(n - 1, -1, -1): #從最後一個數開始n-1,n-2,n-3.........直到0
            self.heapify(nums, i, n, compare) #i值一直換，丟入heapify function裡
        for i in range(n - 1, 0, -1): #從最後面開始跑迴圈(尾，頭，往前推一個)
            nums[0], nums[i] = nums[i], nums[0] #這邊是要排序，拿出去值後，第一個數要和最後一個數互換，繼續排列，直到排完，所以也是跑迴圈
            self.heapify(nums, 0, i, compare)
        return nums

    def compare(a, b):
        if a < b:
            return -1
        if a > b:
            return 1 #這邊設定一個用來比較的參數compare，如果帶入的值a<b，回傳-1，a>b回傳1
        return 0

    def heapify(self, nums ,index, size, compare):
        largest = index #設index為最大父節點
        rightindex = 2 * index + 2 #父節點index的右子節點在位置(2index+2)
        leftindex = 2 * index + 1 #父節點index的左子節點在位置(2index+1)
        if leftindex < size and compare(nums[leftindex], nums[largest]) > 0:
            largest = leftindex #如果左邊的子節點存在，運用compare function表較，>0代表a>b，左邊的子節點則變成父節點
        if rightindex < size and compare(nums[rightindex], nums[largest]) > 0:
            largest = rightindex #如果右邊的子節點存在，運用compare function表較，>0代表a>b，右邊的子節點則變成父節點
        if largest != index:
            nums[largest], nums[index] = nums[index], nums[largest] #如果原本設定的父節點不是最大的，就互換位置
            self.heapify(nums, largest, size, compare)
```
