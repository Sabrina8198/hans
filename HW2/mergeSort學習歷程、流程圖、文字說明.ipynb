# MergeSort
* 一開始的個人理解
  * 1.最初先將數列分成兩半
  * 2.再一直分到剩兩個兩個一組，如果總數為奇數，則剩的數自己一組
  * 3.兩個兩個比大小，小的放左，大的放右
  * 4.合併後比較，'依序'比較
  * 5.比到最後完成排序
  
  * 參考資料:
    * Pseudocode : https://www.slideshare.net/MJabin/merge-sort-and-quick-sort
    * GeeksforGeeks : https://www.geeksforgeeks.org/merge-sort/

```python
class Solution(object):
    def merge_sort(self, nums):
        middle = len(nums)//2
        left = nums[:middle]
        right = nums[middle:]
        self.merge_sort(left)
        self.merge_sort(right)
        
        i = j = k = 0
        for k in range(0, len(nums)):
            if left[i] <= right[j]:
                nums[k] = left[i]
                i+=1
            else:
                nums[k] = right[j]
                j+=1
```

### 轉換成助教要求的格式

```python
class Solution(object):
    def merge_sort(self, nums):
        if len(nums)<2: #設定限制，以防進入無限迴圈
            return nums
        else:
            middle = len(nums)//2
            left = nums[:middle]
            right = nums[middle:] #依照middle，把nums切成兩半
            self.merge_sort(left)
            self.merge_sort(right)
            
            i = j = k = 0
            while i < len(left) and j < len(right):
                if left[i] < right[j]:
                    nums[k] = left[i]
                    i = i+1 #如果左邊的數小於右邊的，將左邊的數加到空的nums的k
                else:
                    nums[k] = right[j]
                    j = j+1 #跟以上同邏輯，右邊的加進去
                k = k+1 #k要跟著移動，加完一個要繼續到下一個位置
            while i < len(left):
                nums[k] = left[i]
                i = i+1
                k = k+1

            while j < len(right):
                nums[k] = right[j]
                j = j+1
                k = k+1
            return nums
```
